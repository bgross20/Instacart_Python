## Creating Bar Charts 

```python
## Creating bar charts
df_ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
```

```python
## Creating bar charts with sort_index function 
df_ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
## Printing frequencies
df_ords_prods_merge['orders_day_of_week'].value_counts()
```

```python
## Printing frequencies with sort_index
df_ords_prods_merge['orders_day_of_week'].value_counts().sort_index()
```

```python
## Creating bar charts with colors 
df_ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar(color = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink'])
```

## Exporting Charts

```python
## Saving as a bar 
bar = df_ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar(color =['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink'])
```

```python
## Exporting data
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```
