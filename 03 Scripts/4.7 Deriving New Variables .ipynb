### Define a function

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

### Apply the function

```python
df['price_range'] = df.apply(price_label, axis =1)
```

```python
## Value Counts 
df['price_range'].value_counts(dropna = False)
```

```python
## Consistency checks 
df['prices'].max()
```

## If-statements with for-loops

```python
## Determine which day of week is busiest 
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
## Create an empty list 
result = []

for value in df_ords_prods_merge['orders_day_of_week']:
    if value  == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else: 
        result.append('Regularly busy')
```

### Question 2 - Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
## Determine which day of week is busiest 
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
## Create an empty list 
result_v2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_v2.append("Busiest days")
  elif value == 4 or value == 3:
    result_v2.append("Slowest days")
  else:
    result_v2.append("Regularly busy")
```

```python
## Merge with dataframe 
df_ords_prods_merge['busiest_days'] = result_v2
```

```python
## Consistency Checks 
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

### Question 4 - When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
## Determine which hour of day is busiest
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

```python
## Create an empty list 
result_v3 = []

for value in df_ords_prods_merge['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result_v3.append('Most orders')
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        result_v3.append('Fewest orders')
    else: 
        result_v3.append('Average orders')
```

```python
## Merge with dataframe 
df_ords_prods_merge['busiest_period_of_day'] = result_v3
```

```python
## Consistency Checks 
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
